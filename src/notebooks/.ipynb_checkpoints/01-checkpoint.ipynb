### EDA on the raw data

```python
import pandas as pd
df = pd.read_csv('../data/census.csv')
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.isnull().sum() ## no missing values
```

```python
df.columns = [i.strip() for i in df.columns.tolist()]
```

```python
def remove_spaces(x):
    return x.strip()
```

```python
cat_cols = df.dtypes[df.dtypes=='object'].index
```

```python
for col in cat_cols:
    df[col] = df[col].map(remove_spaces)
```

```python
df['target'] = df['salary'].map(lambda x: 1 if x=='<=50K' else 0)
```

```python
df.groupby('workclass').agg({'target':['count','mean']})
```

```python
df.groupby('education').agg({'target':['count','mean']})
```

```python
## Drop education-num
df.drop('education-num',axis=1,inplace=True)
```

```python
df.groupby('marital-status').agg({'target':['count','mean']})
```

```python
df.groupby('occupation').agg({'target':['count','mean']})
```

```python
df.groupby('relationship').agg({'target':['count','mean']})
```

```python
df.groupby('race').agg({'target':['count','mean']})
```

```python
df.groupby('sex').agg({'target':['count','mean']})
```

```python
df.groupby('native-country').agg({'target':['count','mean']}) ### too many levels, drop
```

```python
df.drop(['target','native-country'],axis=1,inplace=True)
```

```python
df.to_csv("../data/cleaned.csv",index=False)
```
